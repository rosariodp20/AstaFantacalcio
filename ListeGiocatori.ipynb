```python
# FUNZIONE UNICA PER TUTTI I RUOLI

from openpyxl import load_workbook


def ListaGiocatori():
    
    Listone=load_workbook(filename='Quotazioni_Fantacalcio.xlsx')
    sheet=Listone['Tutti']
    
    Lista_Port=list()
    Lista_Dif=list()
    Lista_Centr=list()
    Lista_Att=list()
    
    i=3
    while i<600:
        B='B'+str(i)
        C='C'+str(i)
        if sheet[B].value=='P':
            Lista_Port.append(sheet[C].value)
        elif sheet[B].value=='D':
            Lista_Dif.append(sheet[C].value)
        elif sheet[B].value=='C':
            Lista_Centr.append(sheet[C].value)
        elif sheet[B].value=='A':
            Lista_Att.append(sheet[C].value)
        i=i+1
    return Lista_Port,Lista_Dif,Lista_Centr,Lista_Att

#Se utilizzo funzione unica ho in output una tupla: nella posizione "0" ho una lista di portieri, nella posizione "1" ho
#una lista di difensori, nella posizione "2" ho una lista di centrocampisti e nella posizione "3" ho una lista di attaccanti

#Nel main dovrò utilizzare questi comandi per salvare le varie liste separatamente
#A=ListaGiocatori()
#Portieri=A[0]
#Difensori=A[1]
#Centrocampisti=A[2]
#Attaccanti=A[3]
```
